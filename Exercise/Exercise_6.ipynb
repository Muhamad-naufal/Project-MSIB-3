```python
# parrent class
class rumus:
  """Bantuan Rumus Bangun Ruang:

  NAME
    Rumus Bangun Ruang

  DESCRIPTION
    Rumus Bangun Ruang
    ==================
    kubus dan balok
  
  CLASES
    kubus
    balok

  FUNCTION
    hitung_volume()->int
    volume()->int
    hitung_luas()->int
    luas()->int
  """

  luas = 0
  volume = 0

  def hitung_volume():
    print("menghitung volume bangun datar")

  def volume(self):
    return self.volume

  def hitung_luas():
    print("menghitung luas bangun datar")

  def luas(self):
    return self.luas

class kubus(rumus):
  def ___init___(self, sisi):
    self.sisi = sisi
  
  def hitung_volume(self):
    self.volume = self.sisi * self.sisi * self.sisi

  def hitung_luas(self):
    self.luas = 6 * self.sisi * self.sisi

class balok(rumus):
  def ___init___(self, panjang, lebar, tinggi):
    self.panjang = panjang
    self.tinggi = tinggi
    self.lebar = lebar
  
  def hitung_volume(self):
    self.volume = self.panjang * self.tinggi * self.lebar

  def hitung_luas(self):
    self.luas = 2 * (self.panjang * self.lebar + self.lebar * self.tinggi + self.lebar * self.tinggi)
```
